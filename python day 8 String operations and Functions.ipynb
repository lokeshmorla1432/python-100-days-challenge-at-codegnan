```python
#Substring index finding
#find()
s="I am Python BatcH studenT"
print(s.find("t"))
print(s.find("t",10))
print(s.find("a"))
print(s.index("s"))
```

```python
#strip - it removes the white space
s="         lokesh     "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

usernname validation -minimum length -8 -atleast one upper case and one lower case letter -special characters('_','@') -it contains digits #password validation



```python
#password validation
 #username and password validator using case check functions
#username
username=input("Enter username: ")
if len(username) >=8 and username[0].isalpha() : 
    for char in username:
        if char.isupper():
            is_upper=True
        elif char.islower():
            is_lower=True
        elif char.isdigit():
            is_digit=True
        elif char in ["_","@"]:
            is_special=True
    if is_upper and is_lower and is_digit and is_special:
        print("Username  valid!")
    else:
        print("Username not valid!")
else:
    print("Username not valid!")

#password 
password=input("Enter password: ")
if len(password) >=8 and password[0].isalpha() : 
    for char in password:
        if char.isupper():
            is_upper=True
        elif char.islower():
            is_lower=True
        elif char.isdigit():
            is_digit=True
        elif char in ["_","@"]:
            is_special=True
    if is_upper and is_lower and is_digit and is_special:
        print("Password  valid!")
    else:
        print("Password  not valid!")
else:
    print("Password not valid!")
```
